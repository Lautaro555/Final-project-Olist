```python
#Split dataframe in 4, sorted by avg income per month
df_low = values.iloc[0:772]
df_mid_low = values.iloc[773:1544]
df_mid_high = values.iloc[1545:2316]
df_high = values.iloc[2317:3090]
```

```python
#split tier 4 into 4 sub-tiers
tier_1d = values.iloc[0:163]
tier_1c = values.iloc[164:324]
tier_1b = values.iloc[325:486]
tier_1a = values.iloc[487:648]
```

```python
#Add identifying column to each DF 
df_low['tier'] = "4"
df_mid_low['tier'] = '3'
df_mid_high['tier'] = '2'
tier_1d['tier'] = '1d'
tier_1c['tier'] = '1c'
tier_1b['tier'] = '1b'
tier_1a ['tier'] = '1a'
```

```python
#This function recieves a dataframe and name of column to be extracted as parameter, returns list of values adjusted to curve
def get_curve (df, column):
    list = df[column].values.tolist()
    normalized_list = []
    max_value = max(list)
    min_value = min(list)
    for i in list:
        x = ((i - min_value)/(max_value - min_value) * 10)
        normalized_list.append(x)
    
    return normalized_list    
```

```python
#Extract income KPI
x_low = get_curve (df_low, 'total_income')
x_mid_low = get_curve(df_mid_low, 'total_income')
x_mid_high = get_curve(df_mid_high, "total_income")
tier_1d_x = get_curve(tier_1d, 'total_income')
tier_1c_x = get_curve(tier_1c,'total_income')
tier_1b_x = get_curve(tier_1b, 'total_income')
tier_1a_X = get_curve(tier_1a, 'total_income')
```

```python
#Add income KPI as new column
df_low['total_income_kpi'] = x_low
df_mid_low['total_income_kpi'] = x_mid_low
df_mid_high['total_income_kpi'] = x_mid_high
tier_1d['total_income_kpi'] = tier_1d_x
tier_1c['total_income_kpi'] = tier_1c_x
tier_1b['total_income_kpi'] = tier_1b_x
tier_1a['total_income_kpi'] = tier_1a_X
```

```python
#Extract review average KPI 
x_low = get_curve (df_low, 'review_avg')
x_mid_low = get_curve(df_mid_low, 'review_avg')
x_mid_high = get_curve(df_mid_high, "review_avg")
tier_1d_x = get_curve(tier_1d, 'review_avg')
tier_1c_x = get_curve(tier_1c,'review_avg')
tier_1b_x = get_curve(tier_1b, 'review_avg')
tier_1a_X = get_curve(tier_1a, 'review_avg')
```

```python
#Add review KPI as new column
df_low['review_avg_kpi'] = x_low
df_mid_low['review_avg_kpi'] = x_mid_low
df_mid_high['review_avg_kpi'] = x_mid_high
tier_1d['review_avg_kpi'] = tier_1d_x
tier_1c['review_avg_kpi'] = tier_1c_x
tier_1b['review_avg_kpi'] = tier_1b_x
tier_1a['review_avg_kpi'] = tier_1a_X
```

```python
#Extract review delivery average KPI 
x_low = get_curve (df_low, 'Delivery_time')
x_mid_low = get_curve(df_mid_low, 'Delivery_time')
x_mid_high = get_curve(df_mid_high, "Delivery_time") 
tier_1d_x = get_curve(tier_1d, 'Delivery_time')
tier_1c_x = get_curve(tier_1c,'Delivery_time')
tier_1b_x = get_curve(tier_1b, 'Delivery_time')
tier_1a_X = get_curve(tier_1a, 'Delivery_time')
```

```python
#Add delivery average KPI as new column
df_low['delivery_avg_kpi'] = x_low
df_mid_low['delivery_avg_kpi'] = x_mid_low
df_mid_high['delivery_avg_kpi'] = x_mid_high
tier_1d['delivery_avg_kpi'] = tier_1d_x
tier_1c['delivery_avg_kpi'] = tier_1c_x
tier_1b['delivery_avg_kpi'] = tier_1b_x
tier_1a['delivery_avg_kpi'] = tier_1a_X
```

```python
df_high = pd.concat([tier_1a, tier_1b, tier_1c, tier_1d])
```

```python
kpi_table = pd.concat([df_low, df_mid_low, df_mid_high, df_high], axis=0)
```

```python
#create function that subtracts delivery_avg_kpi from 10, returns result
def flip_kpi (df, column):
    list = df[column].values.tolist()
    flipped_list = []
    for i in list:
        res = (10 - i)
        flipped_list.append(res)
        
    return flipped_list  
```

```python
temp_list = flip_kpi (kpi_table, 'delivery_avg_kpi')
```

```python
kpi_table['delivery_avg_kpi*'] = temp_list
```

```python
kpi_table['performance_kpi'] = kpi_table[['total_income_kpi', 'review_avg_kpi', 'delivery_avg_kpi*']].mean(axis=1)
```
